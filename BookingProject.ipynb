```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import AgglomerativeClustering

vect = TfidfVectorizer(ngram_range=(1,2), stop_words='english')
X = vect.fit_transform(df['Type'])
clu = AgglomerativeClustering(n_clusters=8).fit(X.toarray())
df['RoomTypes'] = clu.labels_
```

```python
df = df.drop(df.columns[-1], axis = 1)
```

```python
import re

def classify_room(rt: str) -> str:
    s = rt.lower()

    if any(k in s for k in ['dorm', 'bunk', 'pod', 'tent', 'caravan']):
        base = 'Dorm'
    elif 'studio' in s:
        base = 'Studio'
    elif any(k in s for k in ['apartment', 'chalet', 'house', 'home']):
        base = 'Apartment'
    elif 'suite' in s:
        base = 'Suite'
    elif 'family' in s:
        base = 'Family'
    elif any(k in s for k in ['quadruple', '4-bed', 'four-bedroom']):
        base = 'Quadruple'
    elif any(k in s for k in ['triple', '3-bed', 'three-bedroom']):
        base = 'Triple'
    elif any(k in s for k in ['twin']):
        base = 'Twin'
    elif any(k in s for k in ['double', 'queen', 'king', '2-person']):
        base = 'Double'
    elif any(k in s for k in ['single', '1-bed']):
        base = 'Single'
    else:
        base = 'Other'

    if any(k in s for k in ['premium', 'deluxe', 'executive', 'luxury', 'junior', 'superior']):
        return f'Premium {base}'
    elif any(k in s for k in ['standard', 'basic', 'budget']):
        return f'Standard {base}'
    else:
        return base

df['Type_Room'] = df['Type'].apply(classify_room)
```
